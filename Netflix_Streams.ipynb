```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# upload csv into data frame
netflix_data = pd.read_csv("Netflix/titles.csv")
netflix_data.head()
```

```python
total_programs = len(netflix_data)
print(f"There are {total_programs} movies and shows in this dataset...")
```

```python
# create data frame containing only programs wich contain crime in their list of genres
crime_netflix = netflix_data[netflix_data['genres'].str.contains('crime')]
crime_netflix.head()
```

```python
# genre distribution of all netflix programs
genre_counts = netflix_data.groupby(['genres']).count()['title']
genre_counts = genre_counts.sort_values(ascending=False)
genre_counts.head(25)
```

```python
# check length of full data frame (how many titles are included in the set of crime programs)
print(f"There are {len(crime_netflix)} programs on netflix which would fall into the genre of crime.")
```

```python
# check to see if there will be blanks in data frame (compare to length of entire DataFrame)
crime_netflix.count()
```

```python
# sort all programs by their IMDB rating
crime_rating_sort = crime_netflix.sort_values(['imdb_score'],ascending = False)
crime_rating_sort.head(10)
```

```python
# distibution of programs by the year they were released
programs_by_year = crime_netflix.groupby('release_year').count()['title']
programs_by_year = programs_by_year.sort_values(ascending=False)
programs_by_year.head(25)
```

```python
# check distribution of age ratings
crime_age_certif = crime_netflix.groupby('age_certification').count()['title']
crime_age_certif = crime_age_certif.sort_values(ascending=False)
crime_age_certif
```

```python
# see how many crime programs are shows/movies
crime_type = crime_netflix.groupby('type').count()['title']
crime_type
```

```python
# the average rating
avg_rating_crime = crime_netflix['imdb_score'].mean()
print(f"{avg_rating_crime:.2f}")
```

```python
# average runtime for a crime program
avg_runtime = crime_netflix['runtime'].mean()
print(f"Average runtime for a crime program is{avg_runtime : .0f} minutes and {avg_runtime%60 :.0f} seconds")
```

```python
# Top ten production countries for crime programs on netflix
crime_prod_locations = crime_netflix.groupby(['production_countries']).count()['title']
crime_prod_locations = crime_prod_locations.sort_values(ascending=False)
crime_prod_locations.head(20)
```
